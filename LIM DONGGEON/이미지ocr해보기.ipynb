```python
import cv2
import pytesseract
import pandas as pd
import os

# Tesseract 실행 파일의 경로 설정 (시스템에 따라 경로를 조정)
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp')):
            img_path = os.path.join(folder, filename)
            images.append(img_path)
    return images

def ocr_image(image_path):
    image = cv2.imread(image_path)
    if image is None:
        return None
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    text = pytesseract.image_to_string(gray)
    return text

def main(folder_path, output_csv):
    images = load_images_from_folder(folder_path)
    data = []

    for image_path in images:
        print(f'Processing {image_path}')
        text = ocr_image(image_path)
        if text:
            data.append({'filename': os.path.basename(image_path), 'text': text})

    df = pd.DataFrame(data)
    df.to_csv(output_csv, index=False, encoding='utf-8')

if __name__ == '__main__':
    
    folder_path = '/home/data/test_final'  # 이미지 폴더 경로를 지정하세요
    output_csv = 'output.csv'  # 출력 CSV 파일 경로를 지정하세요
    main(folder_path, output_csv)
```

```python
import cv2
import pytesseract
import numpy as np
import os

# Tesseract의 설치 경로를 설정합니다.
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Windows 환경 예시

def resize_image(image, scale_factor):
    # 이미지 해상도를 스케일 팩터만큼 증가
    height, width = image.shape[:2]
    new_dimensions = (int(width * scale_factor), int(height * scale_factor))
    resized_image = cv2.resize(image, new_dimensions, interpolation=cv2.INTER_LINEAR)
    return resized_image

def convert_to_grayscale(image):
    # 이미지를 그레이스케일로 변환
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return gray_image

def sharpen_image(image, sharpness_factor):
    # 기본 샤프닝 커널 정의
    kernel = np.array([[0, -1, 0],
                       [-1, 5 + sharpness_factor, -1],
                       [0, -1, 0]])
    
    # 샤프닝 필터 적용
    sharpened_image = cv2.filter2D(image, -1, kernel)
    cv2.imshow("Sharpened Image", sharpened_image)
    cv2.waitKey(0)
    return sharpened_image

def preprocess_image(image_path):
    # 이미지를 로드합니다.
    image = cv2.imread(image_path)
    if image is None:
        raise FileNotFoundError(f"Image file '{image_path}' not found or unable to load.")
    
    cv2.imshow("Sharpened Image", image)
    cv2.waitKey(0)

    # 이미지 해상도 증가 (2배)
    resized_image = resize_image(image, 2)

    # 그레이스케일로 변환
    gray_image = convert_to_grayscale(resized_image)

    # 샤프닝 (50% 증가)
    sharpened_image = sharpen_image(gray_image, 2.5)  # 50% 증가

    # 전처리된 이미지 저장
    processed_image_path = 'processed_' + image_path
    cv2.imwrite(processed_image_path, sharpened_image)
    return processed_image_path

def extract_text(image_path):
    # 이미지를 불러와서 OCR 수행
    image = cv2.imread(image_path)
    if image is None:
        raise FileNotFoundError(f"Image file '{image_path}' not found or unable to load.")
    text = pytesseract.image_to_string(image, lang='kor')
    return text

def main(image_path):
    # 이미지 전처리
    processed_image_path = preprocess_image(image_path)
    
    # 텍스트 추출
    text = extract_text(processed_image_path)
    print("Extracted Text:")
    print(text)

# 메인 함수 실행
# 메인 함수 실행
"""home = 'G:/내 드라이브/AILAB/CV/Contest/'
image_path = home + 'fe9683317086dbab.jpg'  # 입퇴원 확인서

print(image_path)
image = cv2.imread(image_path)"""
#if image is None: raise FileNotFoundError(f"Image file '{image_path}' not found or unable to load.")
    
# 이미지를 출력합니다.
#cv2.imshow("Image", image)
# 키 입력을 기다립니다.
#cv2.waitKey(0)

#print(image_path)
#text = pytesseract.image_to_string(image_path, lang='kor')
#print(text)
#main(image_path)



def augment_all_images(input_dir): #, output_dir):
    #if not os.path.exists(output_dir):
        #os.makedirs(output_dir)
    i = 0
    for filename in os.listdir(input_dir):
        print(filename)
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            filename = input_dir + '/' + filename
            text = pytesseract.image_to_string(filename, lang='kor')
            print(text)
            #print(input_dir + '/' + filename) #text)


            #input_path = os.path.join(input_dir, filename)
            #image=cv2.imread(input_path)

            #코드 넣기
           

            #output_path = os.path.join(output_dir, filename)
            #cv2.imwrite(output_path, image)
        i += 1
        # print(i)
        if i % 100 == 0 : break                                                      # 100장만 진행하기
        if i % 314 == 0 : print(i)

#augment_all_images('data/train_2_whiteblack', 'data/train_3_correct')
#augment_all_images('data/test_1_whiteblack', 'data/test_2_padding')
augment_all_images('data/test_f_p_b-20C+40') #, 'data/test_final_padding')           # 불러올 파일 위치

```
