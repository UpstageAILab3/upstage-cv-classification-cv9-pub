```python
import cv2
import numpy as np

# 원본 이미지 읽기
image_path = '/home/0fe5dca9a9183702.jpg'
original_image = cv2.imread(image_path)
height, width = original_image.shape[:2]

# 겹칠 이미지 생성
# 원본 이미지의 좌우 반전
flipped_image = cv2.flip(original_image, 1)

# 사이즈를 30%로 줄이기
scale_percent = 30
width_small = int(width * scale_percent / 100)
height_small = int(height * scale_percent / 100)
small_image = cv2.resize(flipped_image, (width_small, height_small))

# 알파값 적용 (투명도 50%)
alpha = 0.5
overlay = np.zeros_like(original_image)
x_offset = width - width_small
y_offset = height - height_small
overlay[y_offset:y_offset+height_small, x_offset:x_offset+width_small] = small_image

# 원본 이미지와 겹친 이미지 합성
blended_image = cv2.addWeighted(original_image, 1, overlay, alpha, 0)

# 결과 이미지 저장
output_path = '/home/0fe5dca9a9183702+.jpg'
cv2.imwrite(output_path, blended_image)

# 결과 이미지 보기
cv2.imshow('Blended Image', blended_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
