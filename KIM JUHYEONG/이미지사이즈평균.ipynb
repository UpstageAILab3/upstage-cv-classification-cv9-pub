```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


# 이미지가 저장된 디렉토리 경로
directory = "cv_challenge/label_change/train"

# 이미지 크기를 저장할 리스트
dimensions = []

# 디렉토리 내의 각 파일을 순회
for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        filepath = os.path.join(directory, filename)
        with Image.open(filepath) as img:
            dimensions.append(img.size)

# numpy 배열로 변환하여 계산
dimensions = np.array(dimensions)

# 평균 크기 계산
mean_width = np.mean(dimensions[:, 0])
mean_height = np.mean(dimensions[:, 1])

print(f"평균 너비: {mean_width}, 평균 높이: {mean_height}")
```

```python
from matplotlib import font_manager, rc

# 한글 폰트 설정 (윈도우 환경)
font_name = 'Malgun Gothic'
rc('font', family=font_name)


# 이미지 크기를 저장할 리스트
widths = []
heights = []

# 디렉토리 내의 각 파일을 순회
for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        filepath = os.path.join(directory, filename)
        with Image.open(filepath) as img:
            width, height = img.size
            widths.append(width)
            heights.append(height)

# numpy 배열로 변환
widths = np.array(widths)
heights = np.array(heights)

# 너비 분포 히스토그램
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.hist(widths, bins=30, color='blue', alpha=0.7)
plt.title('너비 분포 히스토그램')
plt.xlabel('너비')
plt.ylabel('빈도')

# 높이 분포 히스토그램
plt.subplot(2, 2, 2)
plt.hist(heights, bins=30, color='green', alpha=0.7)
plt.title('높이 분포 히스토그램')
plt.xlabel('높이')
plt.ylabel('빈도')

# 너비 분포 박스플롯
plt.subplot(2, 2, 3)
plt.boxplot(widths, vert=False)
plt.title('너비 분포 박스플롯')
plt.xlabel('너비')

# 높이 분포 박스플롯
plt.subplot(2, 2, 4)
plt.boxplot(heights, vert=False)
plt.title('높이 분포 박스플롯')
plt.xlabel('높이')

# 그래프 표시
plt.tight_layout()
plt.show()
```
